```python
import numpy as np 
from matplotlib import pyplot as plt
from scipy.ndimage.filters import gaussian_filter1d
```

```python
model_name = ["cnn"]
file_type = ["mfcc","spec"]
plot_values = {"mfcc":"Mfcc",
				"spec":"Sprectrogram"}

for current_model in model_name:
    plt.figure()
    for current_file_type in file_type:
        path = "../baseline/outputs/saved_models/"+str(current_model)+"/"+str(current_file_type)
#         print(path)
        train_acc = np.load(path+"/data/train_acc.npy")
        train_acc= gaussian_filter1d(train_acc, sigma=1)
        val_acc = np.load(path+"/data/val_acc.npy")
        val_acc= gaussian_filter1d(val_acc, sigma=1)

        plt.plot(np.arange(len(train_acc)),train_acc)
        plt.plot(np.arange(len(val_acc)),val_acc)

    plt.title("CNN Training Accuracy Plot")
    plt.ylabel("Accuracy")
    plt.xlabel("Epochs")
```
