```python
import numpy as np
import librosa
import librosa.display
from matplotlib import pyplot as plt
import skimage
import sys
%matplotlib inline
```

```python
def plot_data(feat):
    plt.figure()
    d = librosa.amplitude_to_db(np.abs(feat), ref=np.max)
    print(d.shape)
    print(mfcc_feat.shape)
    librosa.display.specshow(d, x_axis='time')
    ax = plt.subplot(111)
    ax.get_xaxis().set_visible(False)
    plt.tight_layout()
#     plt.close()
```

```python
path = "../../data/raw_data/wav_files/10_0.wav"
```

```python
y,sr = librosa.load(path)
mfcc_feat = librosa.feature.mfcc(y= y,sr = sr, n_mfcc=20)
spectorgram_feat = librosa.feature.melspectrogram(y=y, sr=sr)
```

```python
plt.title("Audio waveform")
librosa.display.waveplot(y, sr=sr)
```

```python
plot_data(mfcc_feat)
plot_data(spectorgram_feat)
```
