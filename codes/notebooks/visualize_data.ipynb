```python
import librosa
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
import librosa.display 
import skimage

%matplotlib inline
```

```python
file = "../../data/wav_files//10_2.wav"
```

```python
# wav_file = librosa.load(file)
y, sr = librosa.load(file)
```

```python
mfcc_feat = librosa.feature.mfcc(y= y,sr = sr, n_mfcc=20)
```

```python
d = librosa.amplitude_to_db(np.abs(mfcc_feat), ref=np.max)
print(d.shape)
print(mfcc_feat.shape)
d = d * 255
d = d.astype("uint8")
librosa.display.specshow(d, x_axis='time')
ax = plt.subplot(111)
ax.get_xaxis().set_visible(False)
plt.tight_layout()
skimage.io.imsave("mfcc_2.png",d)
```

```python
spectorgram_feat = librosa.feature.melspectrogram(y=y, sr=sr)
```

```python
d = librosa.amplitude_to_db(np.abs(spectorgram_feat), ref=np.max)

print(d.shape)
print(spectorgram_feat.shape)
librosa.display.specshow(d, x_axis='time')
ax = plt.subplot(111)
ax.get_xaxis().set_visible(False)
plt.tight_layout()
d = d * 255
d = d.astype("uint8")
skimage.io.imsave("spec_2.png",d)
```
